```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.dpi'] = 160
mpl.rc('text', usetex=True)
plt.rcParams.update({'font.size': 7})
```

```python
# data : [serial, np(2), np(4)]
data = np.array([[833.91, 561, 353.74], [844, 573, 365]])
a = np.array([1, 2, 4])
font_size=7
```

```python
fig, (ax) = plt.subplots(1, 1, figsize=(7, 4))
ax.plot(a, data[0], '-o', label="Execution Time[s]")
ax.plot(a, data[1], '--o', label="clock Time[s]")
ax.legend()
ax.set_xlabel("No. of Processors", fontsize=font_size)
ax.set_ylabel("Time[s]", fontsize=font_size)
ax.set_xlim(0,5)
ax.set_title("Simulation time for serial and parallel on cluster", fontsize=font_size)
for i, j in zip(a, data[0]):
    ax.annotate("",
                xy=(i, j), xycoords='data',
                xytext=(i-0.3, j), textcoords='data',
                arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

    ax.annotate(str(j), xy=(i-0.55, j-5),xycoords='data', textcoords='data')
    
fig.savefig('PARAL_SIRL.png')
```
